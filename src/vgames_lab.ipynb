```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from data_cleaning import DataCleaning, VGAMES_DATA

df_g = pd.read_excel(VGAMES_DATA)
data = DataCleaning(df_g)

print(data.drop_columns(''))
```

```python
## Rank videogames genres per selling
top_genres = data.get_occcurrences(df_g['Genre'])

genre_list = top_genres.index.tolist()
occurrences_list = top_genres.tolist()

plt.barh(genre_list, occurrences_list)
plt.xlabel('Num of produced games per genre')
plt.ylabel('Genre')
plt.title('Rank the number of produced games')

plt.show()
```

```python
## Rank videogames production per year 
top_year_prod = data.get_occcurrences(df_g['Year'])

year_prod = top_year_prod.index.tolist()
year_prod_occ_list = top_year_prod.tolist()

plt.barh(year_prod, year_prod_occ_list)
plt.xlabel('Num of produced games')
plt.ylabel('Year')
plt.title('Rank videogames production per year')

plt.show()
```

```python
# Platform that sells most videogames overall
top_platform = data.get_occcurrences(df_g['Platform'].astype(str))

plat_list = top_platform.index.to_list()
plat_occ_list = top_platform.to_list()

plt.barh(plat_list, plat_occ_list)
plt.xlabel('Num of produced games')
plt.ylabel('Platform')
plt.title('Rank videogames production per year')

plt.show()
```

# Understand the trend of annual sales of action games

```python
# Plotting time series analysis

df_genre_bar = df_g.loc[df_g['Genre'] == 'Action']
year = df_genre_bar['Year']
g_sales = df_genre_bar['Global_Sales']

df_genre = df_g.loc[df_g['Genre'] == 'Action'].groupby('Year')['Global_Sales'].sum()

concat_data = {'Years': year,
                'Global_Sales': g_sales}
new_df = pd.DataFrame(concat_data)

plt.bar(new_df['Years'], new_df['Global_Sales'])
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.xticks(rotation = 45)
plt.title('Time Series Data')
plt.show()

plt.plot(df_genre)
plt.plot()
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Time Series Data')
plt.show()
```
