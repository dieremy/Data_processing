```python
import matplotlib.pyplot as plt
```

```python
# X * 2 + 1

data = [
    (1, 3),
    (2, 5),
    (3, 7)
]

plt.scatter([x[0] for x in data], [x[1] for x in data])
```

```python
# x * W + B

W = -1
B = 2

predictions = []

for elem in data:
    input_value = elem[0]
    expected_value = elem[1]

    prediction = input_value * W + B
    predictions.append((input_value, prediction))

    print(expected_value, prediction)
    
plt.scatter([x[0] for x in data], [x[1] for x in data])
plt.plot([x[0] for x in predictions], [x[1] for x in predictions])
```

```python
errors = []

for e in range(10):
    print("epoch:", e)
    
    predictions = []
    for elem in data:
        input_value = elem[0]
        expected_value = elem[1]
        
        prediction = input_value * W + B
        error = prediction - expected_value
        errors.append(error)
        predictions.append((input_value, prediction))
        
        W = W - (input_value * error) * 0.1
        B = B - (error * 0.1)
        
        print(expected_value, prediction, error)
        
    plt.scatter([x[0] for x in data], [x[1] for x in data])
    plt.plot([x[0] for x in predictions], [x[1] for x in predictions])
    plt.show()
```
